```python
import pandas as pd

# Specify the path to your CSV file
path = r'C:\Users\Fujitsu\Downloads\Python IBA\SalesRecords.csv'

# Read the CSV file
df = pd.read_csv(path)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# CREATING A VARIABLE COUNT BY REGION
count_by_region = df.Region.value_counts()
count_by_region
```

```python
x = count_by_region.index
y = count_by_region.values
```

```python
count_by_region.plot.line(x=x, y=y)         # d.plot (x=x, y=y, kind="line")
```

```python
count_by_region.plot(x=x, y=y, kind="bar", figsize=(12,6), title="Sales by Region", xlabel="Region", ylabel="Sales", color="violet", grid=False, rot=70, fontsize=12)
```

```python
count_by_region.plot(x=x, y=y, kind="pie", figsize=(12,6), title="Sales by Region", xlabel="Region", ylabel="Sales", color="violet", grid=False, rot=70, fontsize=12)
```

```python
count_by_region.plot(kind="pie", figsize=(6,6), autopct="%1.1f%%", startangle=90, shadow=True, labels=x, colors=["gold", "green", "yellow", "blue", 'orange', 'cyan'])
```

```python
sales_by_region = df.groupby('Region')['Total Revenue'].sum()
sales_by_region
```

```python
sales_by_region.plot(kind="pie", figsize=(6,6), autopct="%1.1f%%", startangle=90, shadow=True, labels=x, colors=["gold", "green", "yellow", "blue", 'orange', 'cyan'])
```

```python
sales_by_region.plot(kind="pie", figsize=(15,15), autopct="%1.2f%%", startangle=90, shadow=True, labels=None, legend=True, colors=["gold", "green", "yellow", "blue", 'orange', 'cyan'])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (15,5))
plt.title('Region wise Summary')
plt.xlabel('Region')
plt.ylabel('Count')

ax = count_by_region.plot(x=x, y=y, kind="line", legend=True, style= 'r--')

plt.xticks(rotation=5, ha='center');
```

```python
plt.figure(figsize = (15,5))
plt.title('Region wise Summary')
plt.xlabel('Region')
plt.ylabel('Count')

ax = count_by_region.plot(x=x, y=y, kind="line", legend=True, style= 'r--')

plt.xticks(rotation=5, ha='center');

# Adding data labels

for i,j in enumerate(y):
    ax.text(i, j, j, ha='left', va='bottom)

plt.plot(x,y, marker = 'o', c = 'blue')
```

```python
plt.figure(figsize = (15,5))
plt.title('Region wise Summary')
plt.xlabel('Region')
plt.ylabel('Total Revenue')

ax = sales_by_region.plot(x=sales_by_region.index, y=sales_by_region.values, kind="bar", legend=False, style='b')

plt.xticks(rotation=5, ha='center');

# Adding data labels

for i,j in enumerate(y):
    ax.text(i, j, j, ha='left', va='bottom')

plt.plot(x,y, marker = 'o', c = 'blue')
```

```python
plt.figure(figsize = (15,5))
plt.title('Region wise Summary')
plt.xlabel('Region')
plt.ylabel('Total Revenue')

ax = sales_by_region.plot(x=sales_by_region.index, y=sales_by_region.values, kind="bar", legend=False, color='b')

plt.xticks(rotation=5, ha='center');

# Adding data labels using text()
for i, j in enumerate(sales_by_region.values):
    plt.text(i, j, format(j, ',.1f'), color='black', fontweight='bold', ha='center', va='bottom')
```

```python
plt.figure(figsize = (15,5))
plt.title('Region wise Summery')
plt.xlabel('Region')
plt.ylabel('Count')

ax = count_by_region.plot(x=x, y=y, kind="line", legend=False, style = 'r--')
plt.xticks(rotation=5, ha='center');

# Adding data labels
for i, j in enumerate(y):
ax.text(i, j, j, ha='left', va='bottom')
plt.plot(x,y, marker = 'o', c = 'blue')
```

```python
plt.figure(figsize = (15,5))
plt.title('Region wise Summery')
plt.xlabel('Region')
plt.ylabel('Count')
ax = count_by_region.plot(x='Region', y='Count', kind="line", legend=False, style='r--')

plt.xticks(rotation=5, ha='center')

# Adding data labels
for i, j in enumerate(y):
    ax.text(i, j, j, ha='left', va='bottom')

plt.plot(x, y, marker='D', c='blue')
plt.show()
```

```python
plt.figure(figsize = (15,5))
plt.title('Region wise Summery')
plt.xlabel('Region')
plt.ylabel('Count')
ax = count_by_region.plot(x='Region', y='Count', kind="barh", legend=False, style='r--')

plt.xticks(rotation=5, ha='center')

# Adding data labels
for i, j in enumerate(sales_by_region):
    ax.annotate(format(j, ',.1f'),
                (j, i),
                ha='left', va='center',
                xytext= (5,0),
                textcoords='offset points')
plt.show()
```

```python
df2 = df.groupby(['Region']).agg(TotalRevenue = ('Total Revenue', 'sum'), \
TotalCost = ('Total Cost', 'sum'), \
TotalProfit = ('Total Profit', 'sum'))
```

```python
ax = df2[['TotalRevenue', 'TotalCost', 'TotalProfit']].plot(kind='bar', legend=True, fig=(6,6))
plt.title('Region wise Summary')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=30, ha='center')
for p in ax.patches:
 ax.annotate("{:,.1f}".format(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')
```
